```python
# Our summary shows skiwness lets plot to see
# Get unique indicators
indicators = df['Subject Descriptor'].unique()

# Plot a histogram for each indicator
for indicator in indicators:
    # Extract the values for this indicator, dropping any nulls
    data = df.loc[df['Subject Descriptor'] == indicator, 'Value']
    
    # Create a new figure for each histogram
    plt.figure()
    plt.hist(data, bins=20)          # 20 bins we can adjust
    plt.title(f'Distribution of {indicator}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)                   # optional grid for readability
    plt.show()
```

```python
#Trend analysis
df['Year']=df['Year'].astype(int)
years = sorted(df['Year'].unique())
import seaborn as sns
# Plot trend for each indicator
indicators = df['Subject Descriptor'].unique()
for indicator in indicators:
    plt.figure(figsize=(10, 5))
    temp = df[df['Subject Descriptor'] == indicator]
    sns.lineplot(data=temp, x='Year', y='Value', hue='Country', legend=False)
    plt.title(f'Trend of {indicator} Over Time')
    plt.xticks(years, rotation=45)
    plt.ylabel(indicator)
    plt.grid(True)
    plt.show()
```

```python
#Let see them by specific countries since difficult to interpret
# let see top 10 (based on average GDP)countries trend analysis
top_10 =df.groupby('Country')['Value'].mean().sort_values(ascending=False).head(10).index
filtered_df = df[df['Country'].isin(top_10)]
indicators = filtered_df['Subject Descriptor'].unique()
years = sorted(df['Year'].unique())
for indicator in indicators:
    plt.figure(figsize=(10, 5))
    temp = filtered_df[filtered_df['Subject Descriptor'] == indicator]
    sns.lineplot(data=temp, x='Year', y='Value', hue='Country')
    plt.title(f'Trend of {indicator} Over Time (Top 10 GDP Countries)')
    plt.xticks(years, rotation=45)
    plt.ylabel(indicator)
    plt.grid(True)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.tight_layout()
    plt.show()
```

```python
years = [2020, 2023, 2025]
pop_n = df[df['Subject Descriptor'] == 'Population']
for year in years:
    subset = pop_n[pop_n['Year'] == year]  # Filter for the specific year
    top5 = subset.nlargest(5, 'Value')[['Country', 'Value']]  # Get the top 5 countries
    print(f"Top 5 Most Populous Countries in {year}:")
    print(top5.to_string(index=False)) 
```

```python
# Filter Population in 2020
pop_2020 = df[(df['Subject Descriptor'] == 'Population') & (df['Year'] == 2020)]

# Select the top 5 by population value
top5_2020 = pop_2020.nlargest(5, 'Value')[['Country', 'Value']]
print(f'Top 5 countries in 2020:')
print(top5_2020.to_string(index=False))
```

```python
#Identify fastest-growing economies based on GDP

gdp_data = df[df['Subject Descriptor'] == 'Gross domestic product, current prices']
gdp_data_sorted = gdp_data.sort_values(by=['Country', 'Year'])
#  Calculate the year-on-year GDP growth rate
gdp_data_sorted['GDP Growth Rate'] = gdp_data_sorted.groupby('Country')['Value'].pct_change() * 100

# Filter out the null values (for the first year, where there's no previous year for comparison)
gdp_growth_data = gdp_data_sorted.dropna(subset=['GDP Growth Rate'])

#  Get the top 5 fastest-growing economies based on GDP growth rate in the latest year
latest_year = gdp_growth_data['Year'].max()
latest_gdp_growth = gdp_growth_data[gdp_growth_data['Year'] == latest_year]

# Top 5 fastest-growing countries by GDP growth rate
top5_fastest_growth = latest_gdp_growth.nlargest(5, 'GDP Growth Rate')[['Country', 'GDP Growth Rate']]

print(top5_fastest_growth)
```

```python
# GDP growth rate for a given country
country_name = 'Ethiopia'
indicator = 'Gross domestic product, current prices'  # Specify the indicator for GDP
country_data = df[(df['Country'] == country_name) & (df['Subject Descriptor'] == indicator)]

# Sort data by year to ensure percentage change is calculated correctly
country_data = country_data.sort_values(by='Year')

# Calculate the GDP growth rate (percentage change between years)
country_data['GDP Growth Rate'] = country_data['Value'].pct_change() * 100

# Get the most recent year
latest_year = country_data['Year'].max()

# Filter the data for the most recent year
recent_data = country_data[country_data['Year'] == latest_year]

# Display the results
print(f"Latest GDP growth rate for {country_name} in {latest_year}:")
print(recent_data[['Year', 'GDP Growth Rate']])
```

```python
# Pivot to get GDP and Population side by side
df_wide = (
    df[df['Subject Descriptor'].isin([
        'Gross domestic product, current prices',
        'Population'
    ])]
    .pivot_table(
        index=['Country','Year'],
        columns='Subject Descriptor',
        values='Value'
    )
    .reset_index()
)

#  Compute GDP per capita
# Since GDP is in billions and Population is in millions,
# the formula is: (GDP in B) / (Population in M)
df_wide['GDP per Capita'] = df_wide['Gross domestic product, current prices'] / df_wide['Population']

# Quick look at 2025 percapita rankings
top_pc_2025 = (
    df_wide[df_wide['Year'] == 2025]
    .nlargest(10, 'GDP per Capita')[['Country','GDP per Capita']]
)
print("Top 10 GDP per Capita in 2025:")
print(top_pc_2025.to_string(index=False))
```

```python
# To compare a specific countries
countries = ['Ethiopia','Egypt','Kenya']

# Filter for those countries
subset_pc = df_wide[df_wide['Country'].isin(countries)]

# View the Year, Country, and GDP per Capita
print(subset_pc[['Country','Year','GDP per Capita']])
```

```python
# specific country and year
countries = ['Ethiopia','Egypt','Kenya']
target_year = 2025


for country_name in countries:
    # Filter your pivoted DataFrame for that country & year
    value = df_wide.loc[(df_wide['Country'] == country_name) &
        (df_wide['Year'] == target_year), 'GDP per Capita']
    
    if not value.empty:
        print(f"{country_name} GDP per Capita in {target_year}: {value.values[0]:.2f} (B/M)")
    else:
        print(f"No data for {country_name} in {target_year}")
```

```python
# #To check if there is correlation b/n population and GDP
# First we can check some Assumptions to interprete the output correctly, Mostly known is linearity and Outliers
plt.figure(figsize=(8, 5))
sns.regplot(data=df_wide, x='Population', y='Gross domestic product, current prices', scatter_kws={'s': 50}, line_kws={'color': 'red'})

plt.title('Population vs GDP with Regression Line')
plt.xlabel('Population (Millions)')
plt.ylabel('GDP (Billions USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#since our data has outliers Pearson correlation may not be the right way for correlation analysis
correlation = df_wide['Population'].corr(df_wide['Gross domestic product, current prices'])
correlation
```

```python
# let's check spearman correlation whch didn't need Linearty,normality and outlier assumptions
from scipy.stats import spearmanr

# Assuming df_wide is your pivoted DataFrame
corr, pval = spearmanr(df_wide['Population'], df_wide['Gross domestic product, current prices'])
print(f"Spearman correlation: {corr:.3f} (p = {pval:.3f})")
```
